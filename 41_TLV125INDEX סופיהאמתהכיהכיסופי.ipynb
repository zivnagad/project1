<p style="color:#00008B; font-weight:bold; font-size:16px;">תשואות שנתיות חריגות גבוהות</p>
ניתן לראות שהקשרים אינם חזקים כמו שחשבנו, לכן בחרנו לבדוק איך המשתנים האלו משפיעים על מניות בעלות תשואה גבוהה מאוד-חריגות:


```python
#חישוב החריגות הגבוהות
quantiles = stock_df['returns per 1 year'].quantile([0.25, 0.5, 0.75])
q1, median_return, q3 = quantiles[0.25], quantiles[0.5], quantiles[0.75]
iqr = q3 - q1
upper_bound = q3 + 1.5 * iqr
lower_bound = q1- 1.5 *iqr

# סינון הנתונים לתשואות חריגות גבוהות ומיון מהגבוה לנמוך
stock_df['company name'] = stock_df['company name'].apply(lambda x: x[::-1])
extremely_high_returns_data = stock_df[stock_df['returns per 1 year'] > upper_bound]
extremely_high_returns_data = extremely_high_returns_data.sort_values(by='returns per 1 year', ascending=False)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(50, 20))

# גרף 1: תשואות מול דיבידנד
sns.barplot(x='company name', y='dividend returns', data=extremely_high_returns_data, color='b', label='Dividend Returns', ax=axs[0])
sns.barplot(x='company name', y='returns per 1 year', data=extremely_high_returns_data, color='yellow', label='Returns per 1 Year', alpha=0.6, ax=axs[0])
axs[0].set_xlabel('Company Name', fontweight='bold', fontsize=30)
axs[0].set_ylabel('Percentage (%)',fontweight='bold', fontsize=30)
axs[0].set_title('Dividend Returns and Returns per 1 Year by Company', fontsize=30, pad=30)
axs[0].legend(fontsize=30)
axs[0].tick_params(axis='x', rotation=90, labelsize=50)
axs[0].tick_params(axis='y', labelsize=50)

# גרף 2: תשואות מול יחס רווח נקי להון עצמי
sns.barplot(x='company name', y='Net profit to equity ratio (%)', data=extremely_high_returns_data, color='b', label='Net Profit to Equity Ratio (%)', ax=axs[1])
sns.barplot(x='company name', y='returns per 1 year', data=extremely_high_returns_data, color='yellow', label='Returns per 1 Year', alpha=0.6, ax=axs[1])
axs[1].set_xlabel('Company Name', fontweight='bold', fontsize=30)
axs[1].set_ylabel('Percentage (%)',fontweight='bold', fontsize=30)
axs[1].set_title('Net Profit to Equity Ratio (%) and Returns per 1 Year by Company', fontsize=30, pad=30)
axs[1].legend(fontsize=30)
axs[1].tick_params(axis='x', rotation=90, labelsize=50)
axs[1].tick_params(axis='y', labelsize=50)

plt.subplots_adjust(wspace=20)
plt.tight_layout()
plt.show()


market_value_by_branch = extremely_high_returns_data.groupby('branch of activity')['market value'].sum().sort_values(ascending=False)
palette = sns.color_palette('Set3', len(market_value_by_branch))
color_map = {branch: color for branch, color in zip(market_value_by_branch.index, palette)}
extremely_high_returns_data['color'] = extremely_high_returns_data['branch of activity'].map(color_map)
extremely_high_returns_data = extremely_high_returns_data.dropna(subset=['color'])

#  גרפי עוגה עבור שווי השוק וענף הפעילות לחברות עם תשואות חריגות גבוהות
fig, axs = plt.subplots(1, 2, figsize=(20, 10))

# גרף עוגה עבור שווי השוק עם צבעים לפי ענף לחברות עם תשואות חריגות גבוהות
axs[0].pie(extremely_high_returns_data['market value'], labels=extremely_high_returns_data['company name'], colors=extremely_high_returns_data['color'], autopct='%1.1f%%', startangle=140)
axs[0].set_title('Market Value Distribution by Company with Branch Colors for High Outliers', fontsize=20)

# גרף עוגה עבור ענף הפעילות לחברות עם תשואות חריגות גבוהות
branch_counts = extremely_high_returns_data['branch of activity'].value_counts()
axs[1].pie(branch_counts, labels=[label for label in branch_counts.index], colors=[color_map[branch] for branch in branch_counts.index], autopct='%1.1f%%', startangle=140)
axs[1].set_title('Branch of Activity Distribution for High Outliers', fontsize=20)

plt.tight_layout()
plt.show()
```

אנו יכולים לראות באופן מפתיע שדווקא החברות בעלות התשואות הגבוהות ביותר (החריגות) לא מחלקות דיבידנדים כלל, או דיבידנד נמוך למשקיעיה.
בנוסף ניתן לראות כי ברוב החברות שעור הרווח הנקי להון העצמי הוא יחסית נמוך, כלומר הרווח שלהן לא גבוה במיוחד.

ניתן לראות כי החברות החריגות עם התשואה הגבוהה ביותר הן דווקא אלו מענפי ההייטק והטכנולוגיה, חברות שנמצאות בצמיחה ומשקיעים מוכנים להגדיל את שווי השוק שלהן מתוך רצון שיתפתחו. זה גם
מסביר את התוצאות שיצאו לנו קודם לכן, חברות מהסוג הזה לא יחלקו דיבידנדים מכיוון שהמשקיעים שלה רוצים שהכסף יושקע בפיתוח ולא בחלוקת "מתנות" למשקיעים, אלו חברות שנמצאות בצמיח.

#### במסקנה זו נרצה לבדוק האם עליה או ירידה בתשואות הענפים מושפעות בתקופות שונות.

```python
unique_branches = stock_df['branch of activity'].unique()

fig, axes = plt.subplots(6, 2, figsize=(20, 30), constrained_layout=True)
axes = axes.flatten()

for i, branch in enumerate(unique_branches):
    branch_df = stock_df[stock_df['branch of activity'] == branch]
    
    average_3_years = branch_df['returns per 3 years'].mean()
    average_1_year = branch_df['returns per 1 year'].mean()
    average_6_months = branch_df['returns per 6 months'].mean()
    average_3_months = branch_df['returns per 3 months'].mean()

    average_returns = pd.DataFrame({
        'Period': ['3 Years', '1 Year', '6 Months', '3 Months'],
        'Average Returns (%)': [average_3_years, average_1_year, average_6_months, average_3_months]
    })

    sns.barplot(x='Period', y='Average Returns (%)', data=average_returns, palette='viridis', ax=axes[i])
    axes[i].set_title(f'Average Returns in {branch}')
    axes[i].set_xlabel('Period')
    axes[i].set_ylabel('Average Returns (%)')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.show()
```

##### סיכונים- הון למאזן:

* יחס גבוה (מעל 0.5 או 50%)
* יחס בטווח הביניים (בין 0.3 ל-0.5)
* יחס נמוך (מתחת ל-0.3 או 30%)

החלטנו לקחת את החברות בעלות סטיית התקן הנמוכה (חברות בטוחות יותר), אנחנו מניחות שכאשר נבדוק את ההון למאזן של אותן החברות נקבל הון למאזן גבוה כי הן חברות בטוחות.

```python
stock_df['std'] = stock_df[returns_columns].std(axis=1)
threshold = stock_df['std'].quantile(0.25)
safe_companies = stock_df[stock_df['std'] <= threshold]
safe_companies = safe_companies.dropna(subset=['Equity Ratio'])
fig, ax = plt.subplots(1, 2, figsize=(18, 8))

sns.barplot(data=safe_companies, x='company name', y='std', ax=ax[0])
ax[0].set_title('low std')
ax[0].set_xlabel('company name')
ax[0].tick_params(axis='x', rotation=90)

sns.barplot(data=safe_companies, x='company name', y='Equity Ratio', ax=ax[1])
ax[1].set_title('Equity Ratio')
ax[1].set_xlabel('company name')
ax[1].tick_params(axis='x', rotation=90)
ax[1].axhline(30, color='red', linestyle='--', linewidth=2, label='ןזאמל ןוה ףס')
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
returns_columns = ['returns per month', 'returns per 3 months', 'returns per 6 months', 'returns per 1 year', 'returns per 3 years']
stock_df[returns_columns] = stock_df[returns_columns].replace({',': '', '%': ''}, regex=True).apply(pd.to_numeric, errors='coerce')

stock_df['std'] = stock_df[returns_columns].std(axis=1)

std_threshold = stock_df['std'].quantile(0.25)

safe_companies = stock_df[stock_df['std'] <= std_threshold]

safe_companies = safe_companies.dropna(subset=['Equity Ratio'])

safe_companies['capital_category'] = pd.cut(safe_companies['Equity Ratio'], bins=[-np.inf, 30, 50, np.inf], labels=['under 30%', ' 30% - 50%', 'up to 50%'])

capital_counts = safe_companies['capital_category'].value_counts().sort_index()

plt.figure(figsize=(8, 5))
sns.barplot(x=capital_counts.index, y=capital_counts.values, palette='viridis')
plt.title('Distribution of companies according to Equity to Balance Sheet')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
